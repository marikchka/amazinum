```python
import pandas as pd
import numpy as np

file_name = 'Energy Indicators.xls'

def read_energy_indicators(file_name):
	energy = pd.read_excel(file_name, header=None, skiprows=17, usecols='C:F', nrows=228)
	energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

	energy.replace('...', np.NaN, inplace=True)
	energy['Energy Supply'] *= 1000000
	energy.loc[0, 'Energy Supply'] = 'Gigajoules'

	return energy

energy = read_energy_indicators(file_name)
energy
```

```python
def change_countries():
	original_energy = energy.copy()
	energy['Country'] = energy['Country'].str.replace(r'\s?\(\S.*\)|\d+', '', regex=True)

	energy['Country'] = energy['Country'].replace({
		'Republic of Korea': 'South Korea',
		'United States of America': 'United States',
		'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
		'China, Hong Kong Special Administrative Region': 'Hong Kong'
	})

	changed_values = original_energy['Country'].compare(energy['Country'])
	return changed_values

print(change_countries())
```

```python
file2_name = 'world_bank.csv'
GPD = pd.read_csv(file2_name, skiprows=4)
GPD
```

```python
def change_countries_GPD():
	original_GPD = GPD.copy()
	GPD['Country Name'] = GPD['Country Name'].replace({
		'Korea, Rep.': 'South Korea',
		'Iran, Islamic Rep.': 'Iran',
		'Hong Kong SAR, China': 'Hong Kong'
	})
	changed_values = original_GPD['Country Name'].compare(GPD['Country Name'])
	return changed_values
print(change_countries_GPD())
```

```python
file3_name = 'Sciamgo Journal and Country Rank.xlsx'
ScimEn = pd.read_excel(file3_name)
ScimEn
```

```python
GPD.rename(columns={'Country Name': 'Country'}, inplace=True)
def answer_one():
	top15 = ScimEn[ScimEn['Rank'] <= 15]
	years = GPD[['Country'] + [str(year) for year in range(2006, 2016)]]

	merged_data = pd.merge(top15, energy, on='Country', how='inner')
	merged_data = pd.merge(merged_data, years, on='Country', how='inner')

	merged_data.set_index('Country', inplace=True)


	return merged_data.head(15)
answer_one()
```

```python
merged = answer_one()
merged.columns
```

```python
def answer_two():
	columns = [str(year) for year in range(2006, 2016)]
	avgGPD = merged[columns].mean(axis=1, skipna=True)
	return avgGPD.sort_values(ascending=False)
answer_two()
```

```python
avgGPD = answer_two()
def answer_three():
	return merged.loc[avgGPD.index[5], '2015']  - merged.loc[avgGPD.index[5], '2006']
answer_three()
```

```python
def answer_four():
	merged['Ratio'] = merged['Self-citations'] / merged['Citations']
	max_ratio = merged['Ratio'].max()
	country_max_ratio = merged['Ratio'].idxmax()
	return (country_max_ratio, max_ratio)
answer_four()
```

```python
def answer_five():
	merged['Estimated Population'] = merged['Energy Supply'] / merged['Energy Supply per Capita']
	estimated_pop = merged['Estimated Population'].sort_values(ascending=False)
	return str(estimated_pop.index[2])
answer_five()
```

```python
def answer_six():
	merged['Citable Doc per Person'] = merged['Citable documents'] / merged['Estimated Population']
	correlation = merged['Citable Doc per Person'].corr(merged['Energy Supply per Capita'])
	return correlation
answer_six()
```

```python
ContinentDict  = {'China':'Asia', 
	'United States':'North America', 
	'Japan':'Asia', 
	'United Kingdom':'Europe', 
	'Russian Federation':'Europe', 
	'Canada':'North America', 
	'Germany':'Europe', 
	'India':'Asia',
	'France':'Europe', 
	'South Korea':'Asia', 
	'Italy':'Europe', 
	'Spain':'Europe', 
	'Iran':'Asia',
	'Australia':'Australia', 
	'Brazil':'South America'}
def answer_seven():
  merged['Continent'] = merged.index.map(ContinentDict)
  df = merged.groupby('Continent')['Estimated Population'].agg(['size', 'sum', 'mean', 'std'])

  return df
answer_seven()
```
