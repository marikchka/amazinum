```python
import os
import numpy as np
import h5py
path = '.'
def load_dataset():
    fn =  os.path.join(path , 'train_signs.h5')
    train_dataset = h5py.File(fn, "r")
    X_train = np.array(train_dataset["train_set_x"][:]) # your train set features
    y_train = np.array(train_dataset["train_set_y"][:]) # your train set labels

    fn =  os.path.join(path , 'test_signs.h5')
    test_dataset =  h5py.File(fn, "r")
    X_test = np.array(test_dataset["test_set_x"][:]) # your test set features
    y_test = np.array(test_dataset["test_set_y"][:]) # your test set labels

    classes = np.array(test_dataset["list_classes"][:]) # the list of classes
    
    y_train = y_train.reshape((1, y_train.shape[0]))
    y_test = y_test.reshape((1, y_test.shape[0]))
    
    return X_train, y_train, X_test, y_test, classes
```

```python
X_train, y_train, X_test, y_test, classes = load_dataset()
y_train = y_train.ravel()
y_test = y_test.ravel()
print ('X_train.shape=', X_train.shape)
print ('X_test.shape=', X_test.shape)
print ('y_train.shape=', y_train.shape)
print ('y_test.shape=', y_test.shape)
```

```python
import matplotlib.pyplot as plt
def display_samples_in_grid(X, n_rows, n_cols= None, y = None ):
    if n_cols is None: 
        n_cols= n_rows
    indices = np.random.randint(0, len(X),n_rows*n_cols)
    for i in range (n_rows): 
        for j in range (n_cols):
            index = n_rows*i+j           
            ax = plt.subplot(n_rows,n_cols,index+1) 
            plt.imshow(X[indices[index]])
            if not (y is None):
                plt.title(y[indices[index]])
            plt.axis('off')

    plt.tight_layout(h_pad=1)
```

```python
plt.figure()

display_samples_in_grid(X_train, n_rows=4, y= y_train)
```

```python
plt.gcf().canvas.manager.set_window_title('Train set')
```

```python
# YOUR_CODE.  Preproces data
# START_CODE 
X_train_flatten = X_train.reshape(X_train.shape[0], -1).T 
X_test_flatten = X_test.reshape(X_test.shape[0], -1).T

X_train_scaled = X_train_flatten / 255.0 
X_test_scaled = X_test_flatten / 255.0
# END_CODE     

print ("number of training examples = " + str(X_train_scaled.shape[1]))
print ("number of test examples = " + str(X_test_scaled.shape[1]))
print ("X_train_scaled shape: " + str(X_train_scaled.shape))

print ("X_test_scaled shape: " + str(X_test_scaled.shape))
```

```python
from sklearn.neural_network import MLPClassifier
# YOUR_CODE.  Train classifier and evaluate the perfromance on train and test sets 
# START_CODE 
clf = MLPClassifier(hidden_layer_sizes=(128, 64), max_iter=500, random_state=20).fit(X_train_scaled.T, y_train)

print("train accuracy= {:.3%}".format(clf.score(X_train_scaled.T, y_train)))
print("test accuracy= {:.3%}".format(clf.score(X_test_scaled.T, y_test)))
# END_CODE 
```

```python
plt.figure()

predicted = clf.predict(X_test_scaled.T)

display_samples_in_grid(X_test, n_rows=4, y= predicted)
```

```python
plt.gcf().canvas.manager.set_window_title('Test set prediction')
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC(C=10, kernel='rbf', gamma=0.001).fit(X_train_scaled.T, y_train)
print("train accuracy= {:.3%}".format(clf.score (X_train_scaled.T, y_train)))
print("test accuracy= {:.3%}".format(clf.score (X_test_scaled.T, y_test)))
```

```python
#conclusion: 
# SVC performed better than MLP, it may be because of high dimensionality, which is hard to MLP
# but easy for SVC due to rbf kernel, its faster convergence, and its fewer tuning requirements
# MLP may be better suited for more complex or larger datasets, if given enough data and proper tuning
```
