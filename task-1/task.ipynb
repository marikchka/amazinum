```python
import random
import matplotlib.pyplot as plt

random_sequence = [random.randint(0, 100) for _ in range(20)] #generate a sequence of 20 random integer values from 0 to 100
 
x = []
y = [88, 26, 71, 56, 34, 93, 4, 53, 87, 97, 23, 18, 54, 48, 78, 40, 70, 89, 4, 59]

#takes indexes of y
for i in range(len(y)):
  x.append(i)

plt.plot(x,y)
plt.show()
```

```python
def find_depth(arr):
  max_depth = 0
  lake = []
  indexes = []
  for i in range(1, len(arr) - 1):
    left_max = max(arr[:i])  #max value on the left (highest peak)(from first el to i)
    right_max = max(arr[i+1:])  #max value on the right (from i to last el)
    
    if left_max > arr[i] and right_max > arr[i]: #enshure that there is no other peak in the lake
      depth = min(left_max, right_max) - arr[i] #water level - bottom
      peak = min(left_max, right_max)
      #update max depth and lake
      if depth > max_depth:  
        max_depth = depth
        #finding indexes of boundaries of lake
        left = arr.index(left_max)
        right = len(arr) - arr[::-1].index(right_max) - 1
        lake = arr[left:right + 1] 
  #creating indexes for plotting red line   
  for i in lake:
    indexes.append(arr.index(i))             

  return max_depth, lake, indexes
      
print(find_depth(y))
```

```python
max_depth, lake, indexes = find_depth(y)
y_for_max_line = [min(lake[0], lake[-1])] * len(x)
y_for_min_line = [min(lake)] * len(x)
plt.plot(x,y) #original data
plt.plot(indexes,lake, color = 'r') #deepest lake
#water level
plt.plot(x, y_for_max_line , ls='--', color='g') 
plt.plot(x, y_for_min_line, ls='--', color='g')

plt.show()
```
